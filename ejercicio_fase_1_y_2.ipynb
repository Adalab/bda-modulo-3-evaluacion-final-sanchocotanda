```python
# cambiar los negativos a nulos y una vez nulos pasarlos a mediana
# Al no haber encontrado relacion directa con otro campo, inputamos la mediana global de la columna salary para que nos perjudique lo menos posible en caso de tener que sacar una media.

# marcar negativos como nulos
df_history.loc[df_history['salary'] < 0, 'salary'] = pd.NA

# imputar NaN con mediana calculada sobre valores válidos
mediana_para_negativos = df_history['salary'].median()
df_history['salary'] = df_history['salary'].fillna(mediana_para_negativos)
```
